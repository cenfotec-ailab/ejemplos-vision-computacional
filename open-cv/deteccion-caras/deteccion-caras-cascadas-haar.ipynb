```python
from matplotlib import pyplot as plt
import cv2
%matplotlib inline
```

```python
# cargar clasificador pre-entrenado
cascada_caras = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
# cargar imagen de prueba
img = cv2.imread('personas.jpg')
# se convierte la imagen en tono gris (b/n)
gris = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# desplegar la imagen
plt.imshow(gris, cmap='Greys_r')
plt.show()
```

```python
# Deteccion de caras
caras = cascada_caras.detectMultiScale(gris, 1.3, 5)

# aqui vamos a almacenar las caras recortadas
recortes = []

# usamos cv2.rectangle para poner un rectangulo a cada
# cara registrada
for (x,y,w,h) in caras:
    # se sobre escribe la image en cada ciclo
    img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),10)
    recortes.append(img[y:y+h, x:x+w])
```

```python
# se despliega la imagen con los bounding boxes
# imagen en color original
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
# deplegar recortes
for recorte in recortes:
    rec = cv2.cvtColor(recorte, cv2.COLOR_BGR2RGB)
    plt.imshow(rec)
    plt.show()
```
